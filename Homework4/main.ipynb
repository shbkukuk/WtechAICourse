```python
from models import model
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data_path = 'data/nn_data.csv'
data = pd.read_csv(data_path,)
data.info()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,[4,7,11]].values
y = data.iloc[:,-1].values
print(y)
X
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.3)
print('x_train shape'+str(x_train.shape))
print('y_train shape'+str(y_train.shape))

x_train= x_train.T
x_test = x_test.T
y_train = y_train.reshape(1,209)
y_test = y_test.reshape(1,90)

print('x_train shape'+str(x_train.shape))
print('y_train shape'+str(y_train.shape))
print('y_train shape'+str(y_test.shape))
```

```python
nn = model.neural_network()
n_x = 3
n_y = 1
n_h = 7 #layer neuron value
layer_dims = (n_x,n_h,n_y)
learning_rate=0.0075


param, costs = nn.fit(x_train,y_train,layer_dims,learning_rate=0.0075,num_iterations=3000)
nn.plot_costs(costs,learning_rate)

```

```python
nn.predict(x_test,y_test,param)
```
